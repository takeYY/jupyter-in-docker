```python
import pandas as pd
import numpy as np
import requests
import json
import time
import random
import sys
sys.path.append("lib.bs4")
from bs4 import BeautifulSoup
import requests
import re

#現在の最大表示行数の出力
pd.set_option('display.max_rows', 100)
#現在の最大表示列数の出力
pd.set_option('display.max_columns', 100)

novels = pd.read_csv('./csv/edogawa_list.csv')
novels
```

```python
list_url = 'https://www.aozora.gr.jp/index_pages/person1779.html'

r = requests.get(list_url)
soup = BeautifulSoup(r.content, 'html.parser')

soup
```

```python
published_novels = soup.body.find_all('ol')[0].find_all('a')

published_novels
```

```python
print(f'novels: {len(novels)}')
print(f'published: {len(published_novels)}')
```

```python
published_novels[0].get('href').replace('../cards/001779/card', '').replace('.html', '')
```

```python
novels.iloc[0].values
```

```python
reader_id_list = []
i = 0
for index, pub_novel in enumerate(published_novels):
    is_valid = False
    pub_name = pub_novel.text
    if pub_name == '随筆銭形平次':
        pub_id = pub_novel.get('href').replace('../cards/001670/card', '').replace('.html', '')
    else:
        pub_id = pub_novel.get('href').replace('../cards/001779/card', '').replace('.html', '')
    name, file_name = novels.iloc[i].values
    name = name.replace('俺は二十面相だ', 'おれは二十面相だ')
    name = name.replace('自作解説（怪人二十面相と少年探偵団）', '自作解説')
    name = name.replace('D', 'Ｄ')
    name = name.replace('Q', 'Ｑ')
    name = name.replace('M', 'Ｍ')
    name = name.replace('R', 'Ｒ')
    if pub_name in ['かいじん二十めんそう']:
        pub_id = file_name.replace("kaijin_nijumenso_", "")
        print(f'except, {index}-{i}: {name},\t\t{pub_id}')
        is_valid = True
        i += 1
    else:
        if pub_name == name:
            print(f'matched,{index}-{i}: {pub_name},\t\t{pub_id}')
            is_valid = True
            i += 1
        else:
            print(f'invalid,{index}-{i}: {pub_name},\t\t{pub_id}')
    if is_valid:
        reader_id_list.append(int(pub_id))

print(reader_id_list)
```

```python
print(f'reader_id_list: {len(reader_id_list)}')
print(f'novels: {len(novels)}')
```

```python
novels['reader_id'] = reader_id_list

novels
```

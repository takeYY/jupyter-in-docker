```python
import pandas as pd
import numpy as np
import requests
import json
import time
import random
import sys
sys.path.append("lib.bs4")
from bs4 import BeautifulSoup
from datetime import datetime as dt
import requests
import re

# 現在の最大表示行数の出力
pd.set_option('display.max_rows', 100)
# 現在の最大表示列数の出力
pd.set_option('display.max_columns', 100)

# 作品リストdf読み込み
novels = pd.read_csv('./csv/edogawa_list.csv')
novels
```

```python
# 青空文庫の江戸川乱歩作品リストページ
list_url = 'https://www.aozora.gr.jp/index_pages/person1779.html'

# アクセスしてhtml読み込み
r = requests.get(list_url)
soup = BeautifulSoup(r.content, 'html.parser')

soup
```

```python
# 公開中の作品リスト取得
published_novels = soup.body.find_all('ol')[0].find_all('a')

published_novels
```

```python
# dfの作品リスト
print(f'novels: {len(novels)}')
# 青空文庫の公開中作品リスト（正確にはn-1）
print(f'published: {len(published_novels)}')
```

```python
# 出版社データを一気に取得
published_date_list = []
i = 0
print(f'[matched] or [invalid], [index]-[i]: [pub_name]-[name],\t\t[publisher_data]')
for index, pub_novel in enumerate(published_novels):
    is_valid = False
    pub_name = pub_novel.text
    if '野村 胡堂' in pub_name:
        continue
    name, file_name = novels[['name', 'file_name']].iloc[i].values
    name = name.replace('俺は二十面相だ', 'おれは二十面相だ')
    name = name.replace('自作解説（怪人二十面相と少年探偵団）', '自作解説')
    name = name.replace('D', 'Ｄ')
    name = name.replace('Q', 'Ｑ')
    name = name.replace('M', 'Ｍ')
    name = name.replace('R', 'Ｒ')
    # 必要なデータの抽出（初版出版日）
    sakuhin_url = f"https://www.aozora.gr.jp/{pub_novel.get('href').replace('../', '')}"
    sakuhin_r = requests.get(sakuhin_url)
    sakuhin_soup = BeautifulSoup(sakuhin_r.content, 'html.parser')
    published_date = sakuhin_soup.find('table', summary='底本データ').find_all('tr')[2].find_all('td')[1].get_text()
    published_date = re.sub('（.*?）', '', published_date).replace('年', '-').replace('月', '-').replace('日', '')
    if 'かいじん二十めんそう' in name:
        pub_id = int(novels['reader_id'].iloc[i])
        if pub_id == 57226 and name == 'かいじん二十めんそう_1':
            is_valid = True
            i += 1
            print(f'matched, {index}-{i}: {pub_name}-{name},\t\t{published_date}')
        if pub_id == 57227 and name == 'かいじん二十めんそう_2':
            is_valid = True
            i += 1
            print(f'matched, {index}-{i}: {pub_name}-{name},\t\t{published_date}')
    else:
        if pub_name == name:
            is_valid = True
            i += 1
            print(f'matched, {index}-{i}: {pub_name}-{name},\t\t{published_date}')
        else:
            print(f'invalid, {index}-{i}: {pub_name}-{name},\t\t{published_date}')
    if is_valid:
        published_date_list.append(dt.strptime(published_date, '%Y-%m-%d'))

print(published_date_list)
```

```python
print(f'teihon_list: {len(published_date_list)}')
print(f'novels: {len(novels)}')
```

```python
novels['published_date'] = published_date_list

novels
```

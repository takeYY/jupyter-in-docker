```python
import pandas as pd
import numpy as np
import requests
import json
import time
import random
import sys
sys.path.append("lib.bs4")
from bs4 import BeautifulSoup
from datetime import datetime as dt
import requests
import re

# 現在の最大表示行数の出力
pd.set_option('display.max_rows', 100)
# 現在の最大表示列数の出力
pd.set_option('display.max_columns', 100)

# 作品リストdf読み込み
novels = pd.read_csv('./csv/edogawa_list.csv')
novels
```

```python
# 青空文庫の江戸川乱歩作品リストページ
list_url = 'https://www.aozora.gr.jp/index_pages/person1779.html'

# アクセスしてhtml読み込み
r = requests.get(list_url)
soup = BeautifulSoup(r.content, 'html.parser')

soup
```

```python
# 公開中の作品リスト取得
published_novels = soup.body.find_all('ol')[0].find_all('a')

published_novels
```

```python
# dfの作品リスト
print(f'novels: {len(novels)}')
# 青空文庫の公開中作品リスト（正確にはn-1）
print(f'published: {len(published_novels)}')
```

```python
# novelsにreader_idを付与
novels_name = list(novels.name)
novels['reader_id'] = 0
for idx, pn in enumerate(published_novels):
    pn_text = pn.get_text()
    if pn_text == '野村 胡堂':
        continue
    pn_id = int( re.findall('card[0-9]+', pn.get('href'))[0].replace('card', '') )
    pn_text = pn_text.replace('おれは二十面相だ', '俺は二十面相だ')
    if pn_id == 57316:
        pn_text = pn_text.replace('自作解説', '自作解説（怪人二十面相と少年探偵団）')
    elif pn_id == 57226:
        pn_text += '_1'
    elif pn_id == 57227:
        pn_text += '_2'
    pn_text = pn_text.replace('Ｄ', 'D')
    pn_text = pn_text.replace('Ｑ', 'Q')
    pn_text = pn_text.replace('Ｍ', 'M')
    pn_text = pn_text.replace('Ｒ', 'R')
    
    novel = novels.query(' name == @pn_text ')
    novels.at[list(novel.index)[0], 'reader_id'] = int(pn_id)

novels
```

```python
# 副題データのリスト
sub_name_list = []
# 初出データのリスト
first_appearance_list = []
# 底本データのリスト
original_text_list = []
# 出版社データのリスト
publisher_list = []
# 出版社データのリスト
published_date_list = []
# 底本の親本かどうかのリスト
is_parent_list = []
i = 0
#print(f'[matched] or [invalid], [index]-[i]: [name],\n\t[初出],\n\t[底本],\n\t[出版社],\n\t[初版発行日]')
#print('-'*100)
for index, pub_novel in enumerate(published_novels):
    is_valid = False
    pub_name = pub_novel.text
    if '野村 胡堂' in pub_name:
        continue
    name, file_name = novels[['name', 'file_name']].iloc[i].values
    name = name.replace('俺は二十面相だ', 'おれは二十面相だ')
    name = name.replace('自作解説（怪人二十面相と少年探偵団）', '自作解説')
    name = name.replace('D', 'Ｄ')
    name = name.replace('Q', 'Ｑ')
    name = name.replace('M', 'Ｍ')
    name = name.replace('R', 'Ｒ')
    # 必要なデータの抽出
    reader_id = pub_novel.get('href').replace('../', '')
    sakuhin_url = f"https://www.aozora.gr.jp/{reader_id}"
    sakuhin_r = requests.get(sakuhin_url)
    sakuhin_soup = BeautifulSoup(sakuhin_r.content, 'html.parser')
    """特定情報の抽出（初出）
    """
    first_appearance = sakuhin_soup.find('table', summary='作品データ')\
                                   .find_all('tr')[1].find_all('td')[1].get_text(',\n\t\t')

    """特定情報の抽出（副題）
    """
    try:
        search_list = sakuhin_soup.find('table', summary='タイトルデータ').find_all('tr')
        for sl in search_list:
            has_sub_name = sl.find_all('td')[0].get_text()
            if has_sub_name == '副題：':
                sub_name = sl.find_all('td')[1].get_text()
                break
            else:
                sub_name = 'なし'
    except:
        sub_name = 'なし'

    """特定情報の抽出（底本）
    """
    original_text_tmp = []
    try:
        original_text = sakuhin_soup.find('table', summary='親本データ')\
                                    .find_all('tr')[0].find_all('td')[1].get_text()
        original_text_tmp.append(original_text)
        is_parent_list.append(True)
    except:
        pass
    original_text = sakuhin_soup.find('table', summary='底本データ')\
                                .find_all('tr')[0].find_all('td')[1].get_text()
    original_text_tmp.append(original_text)
    is_parent_list.append(False)

    """特定情報の抽出（出版社）
    """
    publisher_tmp = []
    try:
        publisher = sakuhin_soup.find('table', summary='親本データ')\
                                .find_all('tr')[1].find_all('td')[1].get_text()
        publisher_tmp.append(publisher)
    except:
        pass
    publisher = sakuhin_soup.find('table', summary='底本データ')\
                            .find_all('tr')[1].find_all('td')[1].get_text()
    publisher_tmp.append(publisher)

    """特定情報の抽出（初版発行日）
    """
    published_date_tmp = []
    try:
        published_date = sakuhin_soup.find('table', summary='親本データ')\
                                     .find_all('tr')[2].find_all('td')[1].get_text()
        published_date = re.sub('（.*?）', '', published_date).replace('年', '-')
        published_date = published_date.replace('月', '') if '日' not in published_date else published_date.replace('月', '-').replace('日', '')
        published_date_tmp.append(published_date)
    except:
        pass
    published_date = sakuhin_soup.find('table', summary='底本データ')\
                                 .find_all('tr')[2].find_all('td')[1].get_text()
    published_date = re.sub('（.*?）', '', published_date).replace('年', '-')
    published_date = published_date.replace('月', '') if '日' not in published_date else published_date.replace('月', '-').replace('日', '')
    published_date_tmp.append(published_date)

    if 'かいじん二十めんそう' in name:
        pub_id = int(novels['reader_id'].iloc[i])
        if pub_id == 57226 and name == 'かいじん二十めんそう_1':
            is_valid = True
            i += 1
        if pub_id == 57227 and name == 'かいじん二十めんそう_2':
            is_valid = True
            i += 1
    else:
        if pub_name == name:
            is_valid = True
            i += 1
    if is_valid:
        #print(f'matched, {index}-{i}: {name}-{sub_name}-,\n\t初出：{first_appearance},\n\t底本：{original_text_tmp},\n\t出版社：{publisher_tmp},\n\t初版発行日：{published_date_tmp}')
        #print('-'*100)
        for fa in first_appearance.split(',\n\t\t'):
            first_appearance_list.append((reader_id, fa))
        sub_name_list.append(sub_name)
        for ot in original_text_tmp:
            original_text_list.append((reader_id, ot))
        for p in publisher_tmp:
            publisher_list.append(p)
        for pub_d in published_date_tmp:
            published_date_list.append(pub_d)
    else:
        #print(f'invalid, {index}-{i}: {name}')
        #print('-'*100)
        pass

print()
print('='*100)
print(f'{"="*50}sub_name{"="*50}')
print(sub_name_list)
print()
print(f'{"="*50}first_appearance{"="*50}')
print(first_appearance_list)
print()
print(f'{"="*50}original_text{"="*50}')
print(original_text_list)
print()
print(f'{"="*50}publisher{"="*50}')
print(publisher_list)
print()
print(f'{"="*50}published_date{"="*50}')
print(published_date_list)
print()
print(f'{"="*50}is_parent{"="*50}')
print(is_parent_list)
```

```python
# 作品DBを作成
cols = {'reader_id': list(novels['reader_id']),
        'name': list(novels['name']),
        'sub_name': sub_name_list,
        'file_name': list(novels['file_name'])}

edogawa_df = pd.DataFrame(cols)
edogawa_df
```

```python
# 底本DBを作成
reader_id_list = []
original_text_true_list = []
for reader_id, ot in original_text_list:
    reader_id_list.append(re.sub('.*?card', '', reader_id).replace('.html', ''))
    original_text_true_list.append(ot)
cols = {'reader_id': reader_id_list,
       'original_text': original_text_true_list,
       'publisher': publisher_list,
       'published_date': published_date_list,
       'is_parent': is_parent_list}
original_text_df = pd.DataFrame(cols)
original_text_df
```

```python
# 作品詳細DBを作成
reader_id_list = []
first_appearance_true_list = []
for reader_id, fa in first_appearance_list:
    reader_id_list.append(re.sub('.*?card', '', reader_id).replace('.html', ''))
    first_appearance_true_list.append(fa)
cols = {'reader_id': reader_id_list,
       'first_appearance': first_appearance_true_list}
edogawa_detail_df = pd.DataFrame(cols)
edogawa_detail_df[:100]
```

```python
### csvとして出力
novels.to_csv('./csv/edogawa_list.csv', index=False, encoding='utf_8_sig')

edogawa_df.to_csv('./csv/edogawa.csv', index=False, encoding='utf_8_sig')
original_text_df.to_csv('./csv/original_text.csv', index=False, encoding='utf_8_sig')
edogawa_detail_df.to_csv('./csv/edogawa_detail.csv', index=False, encoding='utf_8_sig')
```
